# Assignments
## 1. How many transactions are there in total:

```python
# This is another method other than looking as [0] of .shape called on the df. This confirms how many non-null rows there are in each column
df.count().compute()
```

```python
# The below would do the trick too
len(df)
```

## Answer: 284807

## 2. How many transactions are fraud, and how many are not fraud?:

```python
# Counting the number of rows for each class.
df.groupby("Class").count().compute()
```

## Answer: Not fraud - 284315, Fraud - 492

## 3. What is the maximum amount in fraud transactions?:

```python
# Selecting the .max() value from the "Amount" column of the df where the "Class" == 1.
df[df.Class == 1]["Amount"].max().compute()
```

## Answer: $2125.87
